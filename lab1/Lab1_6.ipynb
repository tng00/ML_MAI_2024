8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
cartesian_coords = np.random.rand(10, 2)
print('Декартовы координаты:\n', cartesian_coords)

radius = np.sqrt(cartesian_coords[:, 0]**2 + cartesian_coords[:, 1]**2)
angle = np.arctan2(cartesian_coords[:, 1], cartesian_coords[:, 0])

polar_coords = np.column_stack((radius, angle))
print('\nПолярные координаты:\n', polar_coords)
````

````python
import pandas as pd
````

````python
# Load dataset here
data = pd.read_csv('/content/sample_data/student_perfomance.csv')
````

1. Посчитайте среднее кол-во часов уделяемое на сон студентами

````python
avg_sleep_hours = data['Sleep_Hours'].mean()

print(avg_sleep_hours)
````

2. Сколько процентов студентов посещают внеклассовые активности?

````python
participating_rate = (data['Extracurricular_Activities'] == 'Yes').mean() * 100

print(participating_rate, '%', sep='')
````

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
avg_score_female = data[data['Gender'] == 'female']['Exam_Score'].mean()
avg_score_male = data[data['Gender'] == 'male']['Exam_Score'].mean()

if avg_score_female > avg_score_male:
    print('Да.')
else:
    print('Нет.')
````

4. Выведите 25 и 75 перцентили времени сна студентов

````python
percentile25 = data['Sleep_Hours'].quantile(0.25)
percentile75 = data['Sleep_Hours'].quantile(0.75)

print(f'25-й перцентиль времени сна студентов: {percentile25} часов')
print(f'75-й перцентиль времени сна студентов: {percentile75} часов')
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
avg_score_high_sleep = data[data['Sleep_Hours'] > percentile25]['Exam_Score'].mean()
avg_score_low_sleep = data[data['Sleep_Hours'] < percentile75]['Exam_Score'].mean()

if avg_score_high_sleep > avg_score_low_sleep:
    print('Да.')
else:
    print('Нет.')
````

6. Влияет ли физическая активность на результаты экзамена?

````python
corr = data['Physical_Activity'].corr(data['Exam_Score'])

if np.abs(corr) > 0.3:
    print('Да.')
else:
    print('Нет.')
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
avg_score_very_cool_parents = data[data['Parental_Education_Level'].isin(['College', 'Postgraduate'])]['Exam_Score'].mean()
avg_score_cool_parents = data[~data['Parental_Education_Level'].isin(['College', 'Postgraduate'])]['Exam_Score'].mean()

if avg_score_very_cool_parents > avg_score_cool_parents:
    print('Да.')
else:
    print('Нет.')
````

8. Как влияет участие родителей на результаты экзаменов

````python
data['Parental_Involvement_Code'] = pd.Categorical(data['Parental_Involvement']).codes
corr = data['Parental_Involvement_Code'].corr(data['Exam_Score'])

if corr > 0:
    print('Участие родителей положительно влияет на результаты экзаменов, ', end='')
elif corr < 0:
    print('Участие родителей негативно влияет на результаты экзаменов, ', end='')
else:
    print('Участие родителей не влияет на результаты экзаменов.')

if corr != 0 and corr > 0.3:
  print('причем значительно.')
if corr != 0 and corr <= 0.3:
  print('но не существенно.')
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python
avg_score_high_income = data[data['Family_Income'] == 'High']['Exam_Score'].mean()
avg_score_low_income = data[data['Family_Income'] != 'High']['Exam_Score'].mean()

if avg_score_high_income > avg_score_low_income:
    print('Да.')
else:
    print('Нет.')
````

10. Посчитайте долю студентов сдавших экзамен на более чем 70 баллов в семьях с низким доходом и отсутствием доступа к интернету

````python
passed_students_low_income_no_gpt = data[
    (data['Family_Income'] == 'Low') &
    (data['Internet_Access'] == 'No') &
    (data['Exam_Score'] > 70)
].shape[0]

students_low_income_no_gpt = data[
    (data['Family_Income'] == 'Low') &
    (data['Internet_Access'] == 'No')
].shape[0]

passed_percentage = (passed_students_low_income_no_gpt / students_low_income_no_gpt) * 100
print(passed_percentage, '%', sep='')
````
